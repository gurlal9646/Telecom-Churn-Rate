```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn import datasets
from sklearn.preprocessing import scale
```

```python
data = pd.read_csv("telecom_churn_data.csv")
data.columns = ['state','A/c_len','area_code','phone','int_plan','Vmail_plan','Vmail_msg','Day_mins','Day_calls','Day_charge','Eve_mins','Eve_calls','Eve_charge','Nig_mins','Nig_calls','Nig_charge','int_mins','int_calls','int_charge','Custserv_calls','churn']
data.head()
```

```python
from collections import Counter
Counter(data.Vmail_plan)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
```

```python
data.int_plan = enc.fit_transform(data.int_plan)
data.Vmail_plan = enc.fit_transform(data.Vmail_plan)
data.churn = enc.fit_transform(data.churn)
```

```python
x = data.iloc[:,1:]
y = data.churn
X =x.drop(columns=['phone'])
#from sklearn.preprocessing import scale
#x = scale(x)
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y, random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(max_iter=100000,hidden_layer_sizes=(100),learning_rate_init=0.001)
model.fit (X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predict = model.predict(X_test)
accuracy_score(y_test, y_predict)
```

```python
pd.crosstab(y_test, y_predict)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
pd.crosstab(y_test, y_predict)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```
